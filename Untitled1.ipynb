```python
# Basic parser for catalog_content

def parse_catalog_text(s):
    '''
    Heuristics to extract:
      - item_name (text before first newline or 'Item Name:')
      - value & unit if present like 'Value: 72.0' and 'Unit: Ounce'
      - pack quantity (e.g., 'Pack of 6', '6 Pack', 'Pack of 4')
    Returns dict of parsed fields.
    '''
    out = {"raw": s, "item_name": None, "value": None, "unit": None, "pack_qty": None}
    if not isinstance(s, str):
        return out
    text = s.strip()
    # item name heuristics: after "Item Name:" or before first newline
    m = re.search(r"Item Name:\s*(.+?)(?:\n|$)", text, flags=re.IGNORECASE)
    if m:
        out['item_name'] = m.group(1).strip()
    else:
        out['item_name'] = text.split("\n")[0].strip()
    # Value: NUM or Value: 11.25
    m = re.search(r"Value:\s*([\d\.]+)", text, flags=re.IGNORECASE)
    if m:
        try:
            out['value'] = float(m.group(1))
        except:
            out['value'] = None
    # Unit:
    m = re.search(r"Unit:\s*([A-Za-z0-9\%\s\./°]+)", text, flags=re.IGNORECASE)
    if m:
        out['unit'] = m.group(1).strip()
    # Pack quantity: many variants
    m = re.search(r"(?:Pack of|pack of)\s*(\d+)", text, flags=re.IGNORECASE)
    if m:
        out['pack_qty'] = int(m.group(1))
    else:
        m = re.search(r"(\d+)\s*[-]?\s*pack\b", text, flags=re.IGNORECASE)
        if m:
            out['pack_qty'] = int(m.group(1))
    # also try "Includes 4 boxes" -> capture Includes (\d+)
    if out['pack_qty'] is None:
        m = re.search(r"Includes\s*(\d+)\s*(?:boxes|units|packs|count)\b", text, flags=re.IGNORECASE)
        if m:
            out['pack_qty'] = int(m.group(1))
    return out
```

```python
# Add a few quick derived features
def derive_features(df):
    df = df.copy()
    df['title_len'] = df['item_name'].astype(str).apply(len)
    df['num_tokens'] = df['item_name'].astype(str).apply(lambda x: len(x.split()))
    df['num_numbers'] = df['catalog_content'].astype(str).apply(lambda x: len(re.findall(r'\d+', x)))
    df['has_value'] = df['value'].notnull().astype(int)
    df['has_unit']  = df['unit'].notnull().astype(int)
    df['has_image'] = df['image_link'].notnull() & df['image_link'].astype(str).str.strip().ne("")
    return df
```

```python
demo_feats = derive_features(demo_df2)
print(demo_feats[['item_name','value','unit','pack_qty','title_len','num_tokens','num_numbers','has_value','has_unit','has_image']].head(10))
demo_feats.info()
```

```python

# Step 2a: full dataset parse + derived features

# Apply parsing on full train & test
def parse_full(df):
    parsed = df['catalog_content'].astype(str).apply(parse_catalog_text)
    parsed_df = pd.DataFrame(parsed.tolist())
    df = pd.concat([df.reset_index(drop=True), parsed_df.reset_index(drop=True)], axis=1)
    df = derive_features(df)
    # Fill missing pack_qty with 1 (assume single unit)
    df['pack_qty'] = df['pack_qty'].fillna(1)
    # fill missing value with 0
    df['value'] = df['value'].fillna(0)
    return df

print("Parsing and feature engineering train set...")
train_feats = parse_full(train)
print("Parsing and feature engineering test set...")
test_feats  = parse_full(test)

# Check head
print(train_feats[['item_name','value','unit','pack_qty','title_len','num_tokens','num_numbers','has_value','has_unit','has_image']].head())

# Save features for quick reload
train_feats.to_parquet("train_features.parquet", index=False)
test_feats.to_parquet("test_features.parquet", index=False)
print("Saved train_features.parquet and test_features.parquet")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import lightgbm as lgb
from scipy import sparse
```

```python
# ------------------------------
# Parameters
# ------------------------------
NFOLDS = 5
RANDOM_SEED = 42

# ------------------------------
# Target transform: log1p
# ------------------------------
y = np.log1p(train_feats['price'].values)
```

```python
# ------------------------------
# TF-IDF on catalog_content
# ------------------------------
tfidf = TfidfVectorizer(
    max_features=50000,
    ngram_range=(1,2),
    analyzer='word',
    min_df=5,
    max_df=0.9
)

print("Fitting TF-IDF...")
X_text = tfidf.fit_transform(train_feats['catalog_content'].astype(str))
X_test_text = tfidf.transform(test_feats['catalog_content'].astype(str))
print("TF-IDF shape train:", X_text.shape, "test:", X_test_text.shape)
```

```python
# ------------------------------
# Add numeric features: value, pack_qty, title_len, num_tokens, num_numbers, has_value, has_unit
# ------------------------------
num_cols = ['value','pack_qty','title_len','num_tokens','num_numbers','has_value','has_unit']
X_num = train_feats[num_cols].values
X_test_num = test_feats[num_cols].values
```

```python
# Convert numeric to sparse and hstack
X = sparse.hstack([X_text, sparse.csr_matrix(X_num)], format='csr')
X_test = sparse.hstack([X_test_text, sparse.csr_matrix(X_test_num)], format='csr')

print("Final feature shapes:", X.shape, X_test.shape)
```

```python
# ------------------------------
# 5-fold CV LightGBM
# ------------------------------
kf = KFold(n_splits=NFOLDS, shuffle=True, random_state=RANDOM_SEED)
oof_preds = np.zeros(X.shape[0])
test_preds = np.zeros(X_test.shape[0])

lgb_params = {
    'objective': 'regression',
    'metric': 'rmse',
    'verbosity': -1,
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'seed': RANDOM_SEED
}

for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
    print(f"Fold {fold+1}")
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]
    
    train_data = lgb.Dataset(X_train, label=y_train)
    val_data = lgb.Dataset(X_val, label=y_val, reference=train_data)
    
    model = lgb.train(
        lgb_params,
        train_data,
        num_boost_round=1000,
        valid_sets=[train_data, val_data],
        callbacks=[
            lgb.early_stopping(stopping_rounds=50),
            lgb.log_evaluation(period=100)
        ]
    )
    
    oof_preds[val_idx] = model.predict(X_val, num_iteration=model.best_iteration)
    test_preds += model.predict(X_test, num_iteration=model.best_iteration) / NFOLDS
    
```
